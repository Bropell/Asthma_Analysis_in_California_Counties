```python
# Import dependencies
import pandas as pd
```

```python
# Imoprt raw dataset
asthma_df = pd.read_excel('../Raw_Data/asthma-prevalence-3.xlsx')
asthma_df.head()
```

```python
# Isolate columns & remove spaces from columns
asthma_df2 = asthma_df[['COUNTY', 'YEARS', 'AGE GROUP', 'CURRENT PREVALENCE']]
asthma_df2 = asthma_df2.rename(columns = {"AGE GROUP": "AGE_GROUP", "CURRENT PREVALENCE": "CURRENT_PREVALENCE"})
asthma_df2
```

```python
asthma_df2 = asthma_df2[asthma_df2.AGE_GROUP != "All ages"]
#asthma_df2 = asthma_df2.dropna()
asthma_df2 = asthma_df2.fillna(0)
asthma_df2
```
